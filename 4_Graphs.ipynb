       <p>Graphs (Graph Theory) ek mathematical structure hai jo objects (nodes/vertices) aur connections (edges) ko represent karta hai.<br>Yeh social networks, recommendation systems, knowledge graphs, neural networks, aur AI applications mein use hota hai.</p>

Graph Representation:

    (A) --- (B)
     |       |
     |       |
    (C) --- (D)

Nodes:
- A, B, C, D

Edges:
- (A, B)
- (A, C)
- (B, D)
- (C, D)

- (A), (B), (C), (D) -> Yeh Nodes (Vertices) hain.

- Lines (--- or | ) -> Yeh Edges (Connections) hain.


<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>Undirected Graph</h4>
            <ul>Aisa graph jisme edges ka koi direction nahi hota.
                <br>Dono nodes ek doosre se mutual connection rakhti hain.</ul>
    Example: Facebook Friends Network (Agar A, B ka friend hai to B bhi A ka friend hoga).

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([("Alice", "Bob"), ("Alice", "Charlie"), ("Bob", "David"), ("Charlie", "David")])

# Graph visualize karna
plt.figure(figsize=(4, 4))
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12)
plt.title("Undirected Graph Example")
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>Directed Graph</h4>
            <ul>Aisa graph jisme edges ka direction hota hai.
            <br>One-way connection hoti hai (e.g., Twitter followers).
        </ul>
Example: Agar A â†’ B, iska matlab A, B ko follow karta hai, lekin zaroori nahi ke B bhi A ko follow kare

```python
DG = nx.DiGraph()

DG.add_edges_from([("Alice", "Bob"), ("Bob", "Charlie"), ("Charlie", "Alice"), ("Alice", "David")])

plt.figure(figsize=(4, 4))
nx.draw(DG, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12, arrows=True)
plt.title("Directed Graph Example")
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>Weighted Graph</h4>
            <ul>Aisa graph jisme edges ke saath weights (numeric values) bhi hote hain.
                <br>Weight kisi cost, distance, time, ya importance ko represent karta hai.
        </ul>
Example: Google Maps (Do cities ke darmiyan distance ya travel time).

```python
WG = nx.Graph()

edges = [
    ("A", "B", 4),  # A â†’ B (weight = 4)
    ("A", "C", 2),  # A â†’ C (weight = 2)
    ("B", "C", 5),  # B â†’ C (weight = 5)
    ("B", "D", 10), # B â†’ D (weight = 10)
    ("C", "D", 3)   # C â†’ D (weight = 3)
]

WG.add_weighted_edges_from(edges)

plt.figure(figsize=(5, 5))
pos = nx.spring_layout(WG) 
nx.draw(WG, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12)
edge_labels = {(u, v): d for u, v, d in edges}
nx.draw_networkx_edge_labels(WG, pos, edge_labels=edge_labels, font_size=12)

plt.title("Weighted Graph Example")
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>Cyclic Graph</h4>
            <ul>Aisa graph jisme ek ya zyada cycles (loops) hoti hain, yani koi bhi node apne aap ko direct ya indirect route se dobara reach kar sakti hai.

```python
CG = nx.DiGraph()

edges = [("A", "B"), ("B", "C"), ("C", "A"), ("C", "D"), ("D", "E"), ("E", "C")]
CG.add_edges_from(edges)

plt.figure(figsize=(5, 5))
pos = nx.circular_layout(CG)  # Circular layout
nx.draw(CG, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12, arrows=True)

plt.title("Cyclic Graph Example")
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>Acyclic Graph</h4>
            <ul>Acyclic Graph: Aisa graph jisme koi cycle nahi hoti, yani koi bhi path kisi node tak wapas nahi ja sakta..

```python
AG = nx.DiGraph()

edges = [("A", "B"), ("A", "C"), ("B", "D"), ("C", "D"), ("D", "E")]
AG.add_edges_from(edges)

plt.figure(figsize=(5, 5))
pos = nx.spring_layout(AG)
nx.draw(AG, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12, arrows=True)

plt.title("Acyclic Graph Example (DAG)")
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <ul>7. Trie (Prefix Tree)
            <ul>A special tree used for fast searching of words.
                <br>Used in dictionaries, autocomplete, and search engines.
            </ul>

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

trie = Trie()
words = ["apple", "ape", "bat", "ball"]
for word in words:
    trie.insert(word)

TrieGraph = nx.DiGraph()

def add_edges(node, parent=None, char=""):
    if parent is not None:
        TrieGraph.add_edge(parent, char)
        
    for child_char, child_node in node.children.items():
        add_edges(child_node, char, child_char)
        if child_node.is_end_of_word:
            TrieGraph.add_edge(child_char, f"{child_char} (end)")

add_edges(trie.root)

# Plot the Trie tree
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(TrieGraph)
nx.draw(TrieGraph, pos, with_labels=True, node_size=3000, font_size=12, node_color="lightcoral", edge_color="gray")

plt.title("Trie Tree Example")
plt.show()
```

        <ul>8. Spanning Tree (Minimum Tree Covering All Nodes)
            <ul>Ek connected graph ka ek minimum spanning tree hota hai jo minimum edges ke saath sabhi nodes connect kare.
            </ul>

G = nx.Graph()
edges = [
    ("A", "B", 2),
    ("A", "C", 3),
    ("B", "C", 1),
    ("B", "D", 1),
    ("C", "D", 4)
]
G.add_weighted_edges_from(edges)

# Find the Minimum Spanning Tree using Kruskal's Algorithm
MST = nx.minimum_spanning_tree(G, algorithm="kruskal")

# Plot the original graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Positioning nodes
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=2000, font_size=12)
edge_labels = {(u, v): w for u, v, w in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Original Weighted Graph")
plt.show()

# Plot the Minimum Spanning Tree (MST)
plt.figure(figsize=(8, 6))
nx.draw(MST, pos, with_labels=True, node_color="lightgreen", edge_color="red", node_size=2000, font_size=12)
mst_labels = {(u, v): G[u][v]['weight'] for u, v in MST.edges()}
nx.draw_networkx_edge_labels(MST, pos, edge_labels=mst_labels)

plt.title("Minimum Spanning Tree (MST)")
plt.show()
