#### Date-Time ko store karne ke different formats hote hain.  

| **Type**            | **Example**                     | **Description** |
|---------------------|--------------------------------|----------------|
| **Date (YYYY-MM-DD)** | `"2024-02-02"`                | Sirf **date** store karta hai. |
| **Time (HH:MM:SS)**  | `"14:30:00"`                  | Sirf **time** store karta hai. |
| **Timestamp**        | `"2024-02-02 14:30:00"`       | **Date + Time** ka combination. |
| **Timezone-aware**   | `"2024-02-02 14:30:00+05:30"` | Timezone ke saath timestamp. |
| **Timedelta**        | `"2 days 5 hours"`            | **Duration** ko represent karta hai. |

---


```python
from datetime import datetime
```

```python
now = datetime.now()
print("Current Timestamp:", now)

# Specific Date & Time
dt = datetime(2024, 2, 2, 14, 30, 0)
print("Custom Date-Time:", dt)
```

```python
import pandas as pd

data = {
    "Event": ["Meeting", "Lunch", "Project Submission"],
    "Timestamp": ["2024-02-02 09:00:00", "2024-02-02 13:00:00", "2024-02-02 17:30:00"]
}

df = pd.DataFrame(data)

# Convert Timestamp Column to DateTime Format
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

df
```

```python
import pytz

# Set UTC Timezone
utc_time = datetime.now(pytz.utc)
print("UTC Time:", utc_time)

# Convert UTC to US Timezone
us_time = utc_time.astimezone(pytz.timezone("US/Eastern"))
print("US Eastern Time:", us_time)
```

```python
from datetime import timedelta

start_time = datetime(2024, 2, 2, 9, 0, 0)
end_time = datetime(2024, 2, 2, 17, 30, 0)

# Calculate Difference
duration = end_time - start_time
print("Time Duration:", duration)
```

```python
# Create Date-Time Seies
date_rng = pd.date_range(start="2024-02-01", periods=10, freq="D")
df = pd.DataFrame({"Date": date_rng, "Value": range(10)})
df.set_index("Date", inplace=True)

# Resample Weekly
weekly_avg = df.resample("W").mean()
print(weekly_avg)
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>1 Point Data (Coordinates)</h4>
            <ul>Point data ek specific location ko latitude aur longitude ke coordinates se represent karta hai.
            <ul>Example:
                <ul>Location of a coffee shop: Latitude: 40.748817, Longitude: -73.985428.



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'store': ['Coffee Shop'], 'latitude': [40.748817], 'longitude': [-73.985428]}
df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['longitude'], df['latitude'], color='red', s=100, label='Coffee Shop')
plt.title('Point Data Example (Store Location)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>2 Line Data (Paths or Routes)</h4>
            <ul>Line data connected points ko represent karta hai. Yeh ek continuous path dikhata hai jo multiple locations ko connect karta hai.
            <ul>Example:
                <ul>Flight Route: From New York to Los Angeles.



```python
import geopandas as gpd
from shapely.geometry import LineString
```

```python
line = LineString([(40.748817, -73.985428), (34.052235, -118.243683)])
gdf = gpd.GeoDataFrame(geometry=[line])
gdf
```

```python
gdf.plot()
plt.title('Line Data Example (Flight Route)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>3 Polygon Data (Areas)</h4>
            <ul>Polygon data ek enclosed area ko represent karta hai, jo multiple points se define hota hai.
            <ul>Example:
                <ul>City boundary: Polygon representing the boundary of New York City.



```python
from shapely.geometry import Polygon
```

```python
polygon = Polygon([(0, 0), (0, 1), (1, 1), (1, 0)])
gdf = gpd.GeoDataFrame(geometry=[polygon])
gdf
```

```python
gdf.plot()
plt.title('Polygon Data Example (City Boundary)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

<body style="font-family: Sans-serif;">

            <ul>Raster data ko grid cells (pixels) mein represent kiya jata hai, aur har cell ki value ek specific attribute ko dikhati hai.

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h4>5. Vector Data (Points, Lines, Polygons)
</h4>
            <ul>Vector data represents geographic features using points, lines, and polygons.
            <ul>Example:
                <ul>    City map: Cities as points, roads as lines, and parks as polygons.



```python
from shapely.geometry import Point, LineString, Polygon
```

```python
point = Point(0.5, 0.5)
line = LineString([(0, 0), (1, 1), (2, 2)])
polygon = Polygon([(1, 1), (1, 2), (2, 2), (2, 1)])

gdf = gpd.GeoDataFrame(geometry=[point, line, polygon])
gdf
```

```python
gdf.plot()
plt.title('Vector Data Example (Point, Line, Polygon)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

        <h4>6. Geospatial Time Series Data
</h4>
            <ul>Time series data jo geospatial features ke saath combine hota hai. Har time point pe location-based data hota hai.
            <ul>Example:
                <ul>Air pollution levels over time in different cities.


```python
data = {
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    '2019': [30, 20, 15, 25],
    '2020': [32, 22, 17, 28],
    '2021': [34, 24, 19, 30]
}

df = pd.DataFrame(data)
df
```

```python
df.set_index('city').T.plot(kind='line', marker='o', figsize=(8, 6))
plt.title('Geospatial Time Series Data Example (Temperature)')
plt.xlabel('Year')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.show()
```
